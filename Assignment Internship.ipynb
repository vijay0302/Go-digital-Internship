# Task #6

```python
final_merge['TXN_DATE_TIME'] = pd.to_datetime(final_merge['TXN_DATE_TIME'])
```

```python
print("min date is:-",final_merge['TXN_DATE_TIME'].min())
print("max date is:-",final_merge['TXN_DATE_TIME'].max())
```

```python
#Current date is the max date of the dataset
Date = pd.to_datetime('2021-06-30')
```

```python
past_30_days_data =final_merge.loc[(final_merge['TXN_DATE_TIME'] > '2021-05-31') & (final_merge['TXN_DATE_TIME'] < '2021-06-30')]
past_30_days_data
```

```python
# AVERAGE CLaim of last 30 days
AVERAGE_CLAIM = past_30_days_data.groupby('INSURANCE_TYPE').mean()['CLAIM_AMOUNT'].to_dict()
AVERAGE_CLAIM
```

```python
def Claim_deviation(row):
    if (Date - row['TXN_DATE_TIME'])<pd.Timedelta(days=30):
        return -1
    ClAIM_DEVIATION = AVERAGE_CLAIM[row['INSURANCE_TYPE']] / row['CLAIM_AMOUNT']
    if ClAIM_DEVIATION < 0.5:
        return 1
    else:
        return 0
```

```python
final_merge['CLAIM_DEVIATION'] = final_merge.apply(Claim_deviation,axis=1)
final_merge
```

```python
final_merge.CLAIM_DEVIATION.value_counts()
```

### 9216 claims deviation are less than 0.5, 784 claims transaction have history of less than 30 days
